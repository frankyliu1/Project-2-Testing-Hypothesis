```python
sample_size = 500
incidents_for_year = {}
for year in np.arange(2018, 2023):
    url = (
        f"https://data.sfgov.org/resource/wg3w-h783.csv?"
        f"incident_year={year}&"
        f"$select=incident_category,Latitude,Longitude&"
        f"$where=Latitude%20IS%20NOT%20NULL&"
        f"$limit=50000"
    )
    incidents_for_year[year] = (Table.read_table(url)
                                     .sample(sample_size, 
                                             with_replacement=False))
```

```python
url_2011 = (
    f"https://data.sfgov.org/resource/tmnf-yvry.csv?"
    f"date=2011-01-01T00:00:00.000&"
    f"$select=date,category,y,x&"
    f"$where=y%20IS%20NOT%20NULL&"
    f"$limit=50000")
table_2011 = Table.read_table(url_2011)
table_2011 = table_2011.drop("date")

url_2012 = (
    f"https://data.sfgov.org/resource/tmnf-yvry.csv?"
    f"date=2012-01-01T00:00:00.000&"
    f"$select=date,category,y,x&"
    f"$where=y%20IS%20NOT%20NULL&"
    f"$limit=50000")
table_2012 = Table.read_table(url_2012)
table_2012 = table_2012.drop("date")

url_2013 = (
    f"https://data.sfgov.org/resource/tmnf-yvry.csv?"
    f"date=2013-01-01T00:00:00.000&"
    f"$select=date,category,y,x&"
    f"$where=y%20IS%20NOT%20NULL&"
    f"$limit=50000")
table_2013 = Table.read_table(url_2013)
table_2013 = table_2013.drop("date")

url_2014 = (
    f"https://data.sfgov.org/resource/tmnf-yvry.csv?"
    f"date=2014-01-01T00:00:00.000&"
    f"$select=date,category,y,x&"
    f"$where=y%20IS%20NOT%20NULL&"
    f"$limit=50000")
table_2014 = Table.read_table(url_2014)
table_2014 = table_2014.drop("date")

url_2015 = (
    f"https://data.sfgov.org/resource/tmnf-yvry.csv?"
    f"date=2015-01-01T00:00:00.000&"
    f"$select=date,category,y,x&"
    f"$where=y%20IS%20NOT%20NULL&"
    f"$limit=50000")
table_2015 = Table.read_table(url_2015)
table_2015 = table_2015.drop("date")

url_2016 = (
    f"https://data.sfgov.org/resource/tmnf-yvry.csv?"
    f"date=2016-01-01T00:00:00.000&"
    f"$select=date,category,y,x&"
    f"$where=y%20IS%20NOT%20NULL&"
    f"$limit=50000")
table_2016 = Table.read_table(url_2016)
table_2016 = table_2016.drop("date")

url_2017 = (
    f"https://data.sfgov.org/resource/tmnf-yvry.csv?"
    f"date=2017-01-01T00:00:00.000&"
    f"$select=date,category,y,x&"
    f"$where=y%20IS%20NOT%20NULL&"
    f"$limit=50000")
table_2017 = Table.read_table(url_2017)
table_2017 = table_2017.drop("date")

incidents_for_year[2011] = table_2011.sample(500, with_replacement=False)
incidents_for_year[2012] = table_2012.sample(500, with_replacement=False)
incidents_for_year[2013] = table_2013.sample(500, with_replacement=False)
incidents_for_year[2014] = table_2014.sample(500, with_replacement=True)
incidents_for_year[2015] = table_2015.sample(500, with_replacement=False)
incidents_for_year[2016] = table_2016.sample(500, with_replacement=False)
incidents_for_year[2017] = table_2017.sample(500, with_replacement=True)
```

```python
# A function taken from uszipcode documentation to search for latitude and longitude and outputs the zipcode
from uszipcode import SearchEngine

def lat_lng_to_zip(lat, lng):
    # Initialize the Search Engine 
    search = SearchEngine()

    # Search for the zipcode based on latitude and longitude 
    result = search.by_coordinates(lat, lng, radius = 1, returns = 1)

    # Check if a result is found 
    if result:
        # Extract the zip code from result 
        zipcode = result[0].zipcode

        # Check if the zipcode is within San Francisco (starts with "941")
        if zipcode.startswith("941"):
            return zipcode 
    return "Not in San Francisco"

def convert_zip_2011_2017(table):
    zip_from_latlong_array = make_array()
    for i in np.arange(table.num_rows):
        zip_from_latlong = lat_lng_to_zip(table.column("y").item(i), table.column("x").item(i))
        zip_from_latlong_array = np.append(zip_from_latlong_array, zip_from_latlong)
    converted_table = table.drop("y", "x").with_column("zip code", zip_from_latlong_array)
    return converted_table
```

```python
report_2011 = convert_zip_2011_2017(incidents_for_year[2011])
report_2012 = convert_zip_2011_2017(incidents_for_year[2012])
report_2013 = convert_zip_2011_2017(incidents_for_year[2013])
report_2014 = convert_zip_2011_2017(incidents_for_year[2014])
report_2015 = convert_zip_2011_2017(incidents_for_year[2015])
report_2016 = convert_zip_2011_2017(incidents_for_year[2016])
report_2017 = convert_zip_2011_2017(incidents_for_year[2017])
```

```python
report_2011.to_csv('incidents_for_year_2011.csv')
report_2012.to_csv('incidents_for_year_2012.csv')
report_2013.to_csv('incidents_for_year_2013.csv')
report_2014.to_csv('incidents_for_year_2014.csv')
report_2015.to_csv('incidents_for_year_2015.csv')
report_2016.to_csv('incidents_for_year_2016.csv')
report_2017.to_csv('incidents_for_year_2017.csv')
```

```python
def convert_zip_2018_2022(table):
    zip_from_latlong_array = make_array()
    for i in np.arange(table.num_rows):
        zip_from_latlong = lat_lng_to_zip(table.column("Latitude").item(i), table.column("Longitude").item(i))
        zip_from_latlong_array = np.append(zip_from_latlong_array, zip_from_latlong)
    converted_table = table.drop("Latitude", "Longitude").relabel("incident_category", "category").with_column("zip code", zip_from_latlong_array)
    return converted_table
```

```python
report_2018 = convert_zip_2018_2022(incidents_for_year[2018])
report_2019 = convert_zip_2018_2022(incidents_for_year[2019])
report_2020 = convert_zip_2018_2022(incidents_for_year[2020])
report_2021 = convert_zip_2018_2022(incidents_for_year[2021])
report_2022 = convert_zip_2018_2022(incidents_for_year[2022])
```

```python
report_2018.to_csv('incidents_for_year_2018.csv')
report_2019.to_csv('incidents_for_year_2019.csv')
report_2020.to_csv('incidents_for_year_2020.csv')
report_2021.to_csv('incidents_for_year_2021.csv')
report_2022.to_csv('incidents_for_year_2022.csv')
```
